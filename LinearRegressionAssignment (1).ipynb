1.	What is Regression?

```python
Regression is a machine learning technique used to predict a continuous value based on input variables.
```

2.	What is Linear Regression?

```python
Linear Regression is a method that predicts a value using a straight line relationship between variables.It finds the best fitting line using Two Points.
```

3.	When to use Linear Regression? Explain the equation of a straight line.

```python
We should use linear regression when the target variable is numeric and we want to understand the relationship between the target variable and one or more input variables. It is also useful when we want to make predictions based on the input variables.
Equation of Straight Line is Y = mX + C
Where Y is the target variable(Dependent Variable), X is the input variable(Independent Variable), m is the slope of the line, and C is the intercept. 
```

5.	How is the best fit line chosen?

```python
The best fit line is chosen by minimizing the error between predicted and actual values using least squares method.
```

6.	What is gradient descent, and why is it used? Explain the maths.

```python
Gradient descent is an algorithm used to find the minimal value of a function.In linear regression,it is used to minimize the cost function,which measures the error between predicted and actual values.
we use derivative to find the slope of the cost function and update the parameters in the direction of the negative gradient to minimize the cost function
Gradient Descent uses math derivatives to adjust the line step by step until prediction error becomes minimum
```

7.	What are residuals?

```python
Residual = Actual value − Predicted value
It shows how far the prediction is from the real value.It sshow that the distance is minimized between the predicted and actual values. 
```

8.	What is correlation?

```python
Correlation shows how strongly two variables are related.
```

9.	What is multicollinearity?

```python
Multicollinearity happens when two independent variables are highly correlated with each other.
```

10.	How to detect multicollinearity?

```python
we can use Heatmap to detect multicollinearity. Heatmap shows the correlation between variables, and if two independent variables have a high correlation, it indicates multicollinearity.
we can use VIF (Variance Inflation Factor) to detect multicollinearity. VIF measures how much the variance of the estimated regression coefficients is inflated due to multicollinearity.
VIF should be less than 5.
```

11.	What are the remedies for multicollinearity?

```python
1.Do Not Do Anything.
2.Combine the Independent Variables.
3.Remove one of the correlated independent variables.
4.Principal Component Analysis (PCA) can be used to reduce the dimensionality of the data while retaining most of the variance. 
```

12.	What is the R-Squared Statistics?

```python
R-Squared Statistic is a measure of how well the regression line fits the data. It ranges from 0 to 1, where 0 indicates that the model does not explain any of the variability in the target variable, and 1 indicates that the model explains all of the variability in the target variable. A higher R-squared value indicates a better fit of the model to the data.
```

13.	What is an adjusted R-Squared Statistics?

```python
adjusted R-squared is a modified version of R-squared that takes into account the number of independent variables in the model. 
Adjusted R-squared can be used to compare models with different numbers of independent variables, and a higher adjusted R-squared value indicates a better fit of the model to the data.
```

14.	Why do we use adj R-squared?

```python
Because R² always increases when new features are added, even if they are useless.
Adjusted R² penalizes unnecessary features.
```

15.	Why adj R-squared decreases when we use incompetent variables?



```python
Because it penalizes variables that do not improve the model, reducing the score.
```

16.	How to interpret a Linear Regression model?

```python
we can interpret a Linear regression model by looking at the coefficients of the independent variables. 
The coefficient represents the change in the target variable for a one-unit change in the independent variable, while holding all other variables constant. A positive coefficient indicates a positive relationship between the independent variable and the target variable, while a negative coefficient indicates a negative relationship. The magnitude of the coefficient indicates the strength of the relationship between the independent variable and the target variable. 
```

```python
Bias → error from wrong assumptions (underfitting)
Variance → model too sensitive to data (overfitting)
Tradeoff means balancing both for best performance.
```

```python
OLS (Ordinary Least Squares) finds model parameters by minimizing the sum of squared errors between actual and predicted values.
derivation of fomulae used in OLS Method
Regression Equation
Predicted Value
Residual (Error)
Sum of Squared Errors (SSE)
Partial Derivative w.r.t. Slope (m)
Partial Derivative w.r.t. Intercept (c)
Set Derivatives to Zero
Solve Normal Equations
Obtain Slope Formula
Obtain Intercept Formula
```

25.	What is regularization, and why do we need it?

```python
Regularization reduces overfitting by penalizing large coefficients.
We use regularization to prevent overfitting by adding a penalty term to the cost function.
TYpes of Regularization:
Lasso (L1)
Ridge (L2)
Elastic Net (L1 + L2)
```

26.	Explain Ridge Regression.

```python
Ridge Regresion adds L2 penalty, which shrinks coefficients but does not set them to zero.
```

27.	Explain Lasso Regression.

```python
lasso Regression adds L1 penalty, which can shrink some coefficients to zero, effectively performing feature selection.
```

28.	Explain Elastic Net.

```python
Elastic Net combines L1 and L2 penalties, providing a balance between Ridge and Lasso regression. It can shrink coefficients and perform feature selection simultaneously.
```
